# Gedo Regional API Forecast Pipeline

Validates the live Sybilion API forecasts for Gedo regional water and food prices and checks that the plots were written.

```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'project').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

WATER_JOB_ID = 'fa5a3a7b-b0ab-4355-a43a-0a8c3e7500e2'
FOOD_JOB_ID = '369d1492-0842-4875-99c4-c75ba08a6f4b'

water_dir = PROJECT_ROOT / 'project' / 'data' / 'somalia' / 'water' / 'sybilion' / 'regional' / 'gedo'
food_dir = PROJECT_ROOT / 'somalia' / 'data' / 'sybilion_regional_food' / 'gedo'
```

## Validate Saved API Forecasts

```python
water_forecast = pd.read_csv(water_dir / 'gedo_water_sybilion_forecast.csv', parse_dates=['date'])
food_forecast = pd.read_csv(food_dir / 'gedo_food_sybilion_forecast.csv', parse_dates=['date'])

water_plot = water_dir / 'gedo_water_sybilion_forecast.png'
food_plot = food_dir / 'gedo_food_sybilion_forecast.png'

assert len(water_forecast) > 0
assert len(food_forecast) > 0
assert water_forecast['forecast'].notna().all()
assert food_forecast['forecast'].notna().all()
assert water_plot.exists()
assert food_plot.exists()

print('water job:', WATER_JOB_ID)
print('food job:', FOOD_JOB_ID)
print('water plot:', water_plot)
print('food plot:', food_plot)

display(water_forecast)
display(food_forecast)
```
